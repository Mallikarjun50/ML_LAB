```python
from google.colab import files
import pandas as pd

# This will open a file picker for you to upload your CSV
uploaded = files.upload()
```

```python
if uploaded:
    # Get the name of the first uploaded file (assuming only one is uploaded)
    uploaded_filename = next(iter(uploaded))
    print(f"Uploaded file: {uploaded_filename}")

    # Read the CSV file into a pandas DataFrame
    df_uploaded = pd.read_csv(uploaded_filename)

    # Display the first few rows of the DataFrame
    print("First 5 rows of the uploaded DataFrame:")
    print(df_uploaded.head())
else:
    print("No file was uploaded.")
```

```python
print("Missing values before cleaning:")
print(df_uploaded.isnull().sum())
```

```python
print(f"Original DataFrame shape: {df_uploaded.shape}")

df_cleaned = df_uploaded.dropna()

print(f"DataFrame shape after dropping rows with missing values: {df_cleaned.shape}")
print("First 5 rows of the cleaned DataFrame:")
print(df_cleaned.head())
print(df_cleaned)
```

```python
import pandas as pd

# Identify categorical columns from the DataFrame.
# Since df_cleaned only contains 'year' and 'per capita income (US$)', there are no categorical columns to encode.
categorical_cols = [col for col in ['Gender', 'CLASS'] if col in df_cleaned.columns]

# Apply one-hot encoding only if there are categorical columns
if categorical_cols:
    df_encoded = pd.get_dummies(df_cleaned, columns=categorical_cols, drop_first=True)
    print("DataFrame after one-hot encoding:")
    print(df_encoded.head())
else:
    df_encoded = df_cleaned.copy()
    print("No categorical columns found for one-hot encoding.")
    print("Original DataFrame head:")
    print(df_encoded.head())
```

```python
import numpy as np

# Select only numerical columns for outlier detection
numerical_cols = df_encoded.select_dtypes(include=np.number).columns

df_no_outliers = df_encoded.copy()

print(f"Original DataFrame shape: {df_encoded.shape}")

# Loop through numerical columns to detect and remove outliers using IQR
for col in numerical_cols:
    Q1 = df_no_outliers[col].quantile(0.25)
    Q3 = df_no_outliers[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out outliers for the current column
    df_no_outliers = df_no_outliers[(df_no_outliers[col] >= lower_bound) & (df_no_outliers[col] <= upper_bound)]

print(f"DataFrame shape after outlier handling: {df_no_outliers.shape}")
print("First 5 rows of DataFrame after outlier handling:")
print(df_no_outliers.head())
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np

# Select only numerical columns for scaling
numerical_cols = df_no_outliers.select_dtypes(include=np.number).columns

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Apply Min-Max Scaling to the numerical columns
df_scaled = df_no_outliers.copy()
df_scaled[numerical_cols] = scaler.fit_transform(df_scaled[numerical_cols])

print("DataFrame after Min-Max Scaling (first 5 rows):")
print(df_scaled.head())
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Select only numerical columns for scaling
numerical_cols = df_no_outliers.select_dtypes(include=np.number).columns

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply Standard Scaling to the numerical columns
df_standard_scaled = df_no_outliers.copy()
df_standard_scaled[numerical_cols] = scaler.fit_transform(df_standard_scaled[numerical_cols])

print("DataFrame after Standard Scaling (first 5 rows):")
print(df_standard_scaled.head())
```
